```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''Assumes csv file contains header'''
IMUWPower   = "TestingData\\withpower.csv"
IMUWTorque  = "TestingData\\withTorque.csv"
IMUPos      = "TestingData\\poscontrol.csv"
IMUVel      = "TestingData\\velcontrol.csv"

servoPos    = "TestingData\\servoPosTest.csv"
servoVel    = "TestingData\\servoVelTest.csv"

data_frame = pd.read_csv(IMUWPower, sep = ', ', header=0)
data_frame.set_index('time').plot(subplots=True)


data_frame = pd.read_csv(IMUWTorque, sep = ', ', header=0)
data_frame.set_index('time').plot(subplots=True)


data_frame = pd.read_csv(servoPos, sep = ', ', header=0)
data_frame.set_index('time').plot(subplots=True)

data_frame = pd.read_csv(IMUPos, sep = ', ', header=0)
data_frame.set_index('time').plot(subplots=True)

data_frame = pd.read_csv(IMUVel, sep = ', ', header=0)
data_frame.set_index('time').plot(subplots=True)

data_frame = pd.read_csv(servoVel, sep = ', ', header=0)
data_frame.set_index('time').plot(subplots=True)

plt.show()
```
